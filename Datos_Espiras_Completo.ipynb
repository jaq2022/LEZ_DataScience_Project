```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import time
```

```python
# Numero de filas por espira.
rpe = 1023395

# Numero de espiras.
ne = 200

# Labels columnas.
col_names = ["Intensidad", "FechaHora", "Espira"]
```

```python
# Dataframe vacio.
df = pd.DataFrame(columns = ["FechaHora", "Intensidad", "Espira"])
```

```python
# Se leen los datos de manera iterativa.
# Cada iteración lee las filas de una espira completa.
# Después se hacen las agregaciones pertinentes para tener intensidad media diaria de trafico.
start = time.time()
for i in range(0, ne):
    print("Espira ", i + 1)
    data = pd.read_csv("resultados.csv", nrows = rpe, skiprows = i * rpe, header = None, 
                                     delimiter = ";", names = col_names)
    data["FechaHora"] = pd.to_datetime(data["FechaHora"])
    aux = data.groupby(data['FechaHora'].dt.date).agg({"Intensidad": "sum", "Espira": "first"}).reset_index()
    aux["Intensidad"] = (aux["Intensidad"] / 60).round()
    df = pd.concat([df, aux], ignore_index = True)
end = time.time()
ellapsed_time = end - start
print(ellapsed_time)
```

```python
# Se exporta a csv
df.to_csv("200espiras.csv")
```
